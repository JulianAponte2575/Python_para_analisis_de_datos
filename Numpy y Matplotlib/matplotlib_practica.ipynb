# GRAFICOS ELEMENTALES CON MATPLOTLIB

```python
# IMPORTAR PANDAS
import pandas as pd
```

```python
# LISTAS CON DATOS DE REDES SOCIALES 

fbk = ['Facebook', 2449, True, 2006]
twt = ['Twitter', 339, False, 2006]
ig = ['Instagram', 1000, True, 2010]
yt = ['YouTube', 2000, False, 2005]
lkn = ['LinkedIn', 663, False, 2003]
wsp = ['WhatsApp', 1600, True, 2009]

# LISTA CON LOS DATOS
lista_rrss = [fbk, twt, ig, yt, lkn, wsp]

# CREAR DF CON LAS LISTAS 

df = pd.DataFrame(lista_rrss, columns = ['Nombre', 'Cantidad', 'ES_FBK', 'Año'])
df
```

```python
# IMPORTAR MATPLOTLIB

import matplotlib.pyplot as plt
```

```python
# GRAFICO DE LINEAS DE NÚMEROS AL AZAR 

x = [1,2,3,4,5]
y = [1,8,27,64,125]

plt.plot(x,y);
```

### GRÁFICOS PARA LAS REDES SOCIALES

```python
# GRAFICO DE LINEAS PARA CANTIDAD DE USUARIOS

plt.plot(df["Nombre"],df["Cantidad"], 'bo--', linewidth=2, markersize=6)
plt.ylabel("Millones de usuarios")
plt.title("Redes sociales por usuarios");
```

```python
# MISMOS DATOS PERO CON PUNTOS

plt.scatter(df["Nombre"],df["Cantidad"])
plt.ylabel("Millones de usuarios")
plt.title("Redes sociales por usuarios");
```

```python
# MISMO GRÁFICO DE PUNTOS PERO CON MÉTODOS DE PANDAS 

df.plot(kind="scatter",x="Nombre",y="Cantidad")
plt.title("Redes sociales por usuarios");
```

### GRÁFICOS DE BARRAS

```python
# GRAFICO DE BARRAS 

plt.bar(df["Nombre"],df["Cantidad"])
plt.ylabel("Millones de usuarios")
plt.title("Redes sociales por usuarios");
```

```python
# GRAFICO DE BARRAS ORDENADO

df_ordenado = df.sort_values("Cantidad", ascending = False)
plt.bar(df_ordenado["Nombre"],df_ordenado["Cantidad"])
plt.ylabel("Millones de usuarios")
plt.title("Redes sociales por usuarios");
```

```python
# GRÁFICO DE BARRAS CON COLORES, LISTA DE COLORES 

# b: blue.
# g: green.
# r: red.
# c: cyan.
# m: magenta.
# y: yellow.
# k: black.
# w: white.

plt.bar(df_ordenado["Nombre"], df_ordenado["Cantidad"],
       color = ["c","b","m","g","r","b"])
plt.title("Redes sociales por usuarios");
```

```python
# GRAFICO DE PASTEL 

plt.pie(df_ordenado["Cantidad"], labels=df_ordenado["Nombre"])
plt.title("Proporción de usuarios por red social");
```

```python
lista_colores = ['#faed6b', '#FF0000', '#fa6bdc', '#05964b', '#0e76a8', '#d5a1f2']
```

```python
# GRÁFICO DE PASTEL USANDO MATPLOTLIB CON ETIQUETAS

plt.pie(df_ordenado["Cantidad"],
       labels = df_ordenado["Nombre"],
       colors = lista_colores,
       autopct = "%1.1f%%")
plt.title("Proporción de usuarios por red social");
```
