## ANÁLISIS DE DATOS CON MATPLOTLIB

```python
# IMPORTAR LIBRERÍAS

import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
# TRAER LAS TABLAS

ruta = "C:/Users/Julian/Documents/Ejecicios del botcamp/titanic.csv"
ruta_1 = "C:/Users/Julian/Documents/Ejecicios del botcamp/Ventas_Videojuegos.xlsx"

df_titanic = pd.read_csv(ruta)
df_juegos = pd.read_excel(ruta_1)
df_titanic
```

```python
df_juegos
```

```python
# NÚMERO DE HOMBRES

hombres = len(df_titanic[df_titanic["Sex"] == "male"])
hombres
```

```python
# NÚMERO DE MUJERES

mujeres = len(df_titanic[df_titanic["Sex"] == "female"])
mujeres
```

```python
# PROMEDIO DE HOMBRES Y MUJERES DEL BARCO

promedio_hombres = (hombres / len(df_titanic)* 100)
promedio_mujeres = (mujeres / len(df_titanic) * 100)
valores = {"Proporcion_hombres": promedio_hombres,
          "Proporcion_mujeres": promedio_mujeres}
promedios = pd.DataFrame(valores, index=[0])
promedios
```

```python
# EXTRAER DATOS DE ACUERDO A CONDICIOENS

mayores_30 = df_titanic[(df_titanic["Sex"] == "male") & (df_titanic["Age"] > 30)]["Name"].count()
menores_18 = df_titanic[(df_titanic["Sex"] == "male") & (df_titanic["Age"] < 18)]["Name"].count()
mayores_50 = df_titanic[(df_titanic["Sex"] == "male") & (df_titanic["Age"] > 50)]["Name"].count()
```

```python
# GRAFICAR CANTIDADES

d = {'Edad > 30': mayores_30,
    'Edad < 18': menores_18,
    'Edad > 50': mayores_50}

# Crear el gráfico de barras
grafos = plt.bar(d.keys(), d.values(), color=['blue', 'orange', 'green'])
# Agregar etiquetas de datos automáticamente
plt.bar_label(grafos, labels=d.values(), label_type='edge', fontsize=10, color='black')
plt.title("Conteo de hombres por Grupo de Edad");
```

```python
# EXTRAER DATOS DE ACUERDO A CONDICIOENS

mmayores_30 = df_titanic[(df_titanic["Sex"] == "female") & (df_titanic["Age"] > 30)]["Name"].count()
mmenores_18 = df_titanic[(df_titanic["Sex"] == "female") & (df_titanic["Age"] < 18)]["Name"].count()
mmayores_50 = df_titanic[(df_titanic["Sex"] == "female") & (df_titanic["Age"] > 50)]["Name"].count()
```

```python
# GRAFICAR CANTIDADES
datos = {'Edad > 30': mmayores_30,
    'Edad < 18': mmenores_18,
    'Edad > 50': mmayores_50}

# Crear el gráfico de barras
grafo = plt.bar(datos.keys(), datos.values(), color=['blue', 'orange', 'green'])
# Agregar etiquetas de datos automáticamente
plt.bar_label(grafo, labels=datos.values(), label_type='edge', fontsize=10, color='black')
plt.title("Conteo de mujeres por Grupo de Edad");
```

```python
# PROMEDIO DE VENTAS EN JAPON 

promedio_ventas_JP = df_juegos["Ventas JP"].mean()
promedio_ventas_JP
```

```python
# PROMEDIO DE VENTAS NA 

promedio_ventas_NA = df_juegos["Ventas NA"].mean()
promedio_ventas_NA
```

```python
# GRAFICAR VENTAS DE TODOS LOS PAISES

ventas_NA = df_juegos["Ventas NA"].sum().round(2)
ventas_JP = df_juegos["Ventas JP"].sum().round(2)
ventas_EU = df_juegos["Ventas EU"].sum().round(2)
ventas_otros = df_juegos["Ventas Otros"].sum().round(2)
ventas_global = df_juegos["Ventas Global"].sum().round(2) 

v = {
    'ventas NA': ventas_NA,
    'ventas JP': ventas_JP,
    'ventas EU': ventas_EU,
    'ventas otros': ventas_otros,
    'ventas globales': ventas_global
}

# grafico
ventas = plt.bar(v.keys(), v.values(), color=['blue', 'orange', 'green'])
plt.bar_label(ventas, labels=v.values(), label_type='edge', fontsize=10, color='black')
plt.title('Ventas por paises')
plt.xticks(rotation = 10);
```

```python
# GRAFICAR VENTAS DE TODOS LOS PAISES DE FORMA ASCENDENTE

ventas_NA = df_juegos["Ventas NA"].sum().round(2)
ventas_JP = df_juegos["Ventas JP"].sum().round(2)
ventas_EU = df_juegos["Ventas EU"].sum().round(2)
ventas_otros = df_juegos["Ventas Otros"].sum().round(2)
ventas_global = df_juegos["Ventas Global"].sum().round(2) 

v2 = {
    'ventas NA': ventas_NA,
    'ventas JP': ventas_JP,
    'ventas EU': ventas_EU,
    'ventas otros': ventas_otros,
    'ventas globales': ventas_global
}

# grafico
plt.bar(v2.keys(), v2.values(), color=['blue', 'orange', 'green'])
plt.bar_label(ventas, labels=v2.values(), label_type='edge', fontsize=10, color='black')
plt.title('Ventas por paises')
plt.xticks(rotation = 10);
```

```python
# GRAFICAR VENTAS DE TODOS LOS PAISES DE FORMA ASCENDENTE

ventas_NA = df_juegos["Ventas NA"].sum().round(2)
ventas_JP = df_juegos["Ventas JP"].sum().round(2)
ventas_EU = df_juegos["Ventas EU"].sum().round(2)
ventas_otros = df_juegos["Ventas Otros"].sum().round(2)
ventas_global = df_juegos["Ventas Global"].sum().round(2) 

asc = {
    'ventas globales': ventas_global,
    'ventas NA': ventas_NA,
    'ventas EU': ventas_EU,
    'ventas JP': ventas_JP,
    'ventas otros': ventas_otros
}

# GRAFICO
G = plt.bar(asc.keys(), asc.values(), color=['blue', 'orange', 'green'])
plt.bar_label(G, labels = asc.values(), label_type='edge', fontsize=10, color='black')
plt.title('Ventas por paises')
plt.xticks(rotation = 10);
```
