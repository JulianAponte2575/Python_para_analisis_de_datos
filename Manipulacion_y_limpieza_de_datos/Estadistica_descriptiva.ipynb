```python
# IMPORTAR LIBRERIAS 

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# CRERACIÓN DE DF

ruta = "C:/Users/Julian/Documents/Ejecicios del botcamp/Ventas_Videojuegos.xlsx"
df = pd.read_excel(ruta)
df
```

```python
# SEPARACIÓN DE DATOS POR NES Y WII

nes = df[df["Plataforma"] == "NES"]
wii = df[df["Plataforma"] == "Wii"]
```

```python
# VER LA TABLA DE NES

nes.head()
```

```python
# VER LATABLA DE WII

wii.head()
```

```python
# CANTIDADES DE JUEGOS POR NES Y WII

c_nes = len(nes)
c_wii = len(wii)
c_nes, c_wii
```

```python
# DATOS VENTAS JP POR NES Y WII

min_JP_NES = nes["Ventas JP"].min()
max_JP_WII = wii["Ventas JP"].max()
max_JP_NES = nes["Ventas JP"].max()
min_JP_WII = wii["Ventas JP"].min()
```

```python
# DATOS NA NES Y WII

min_NA_NES = nes["Ventas NA"].min()
max_NA_WII = wii["Ventas NA"].max()
max_NA_NES = nes["Ventas NA"].max()
min_NA_WII = wii["Ventas NA"].min()
datos = {
    'Plataforma': ['NES', 'WII'],
    'Ventas JP Min': [min_JP_NES, min_JP_WII],
    'Ventas JP Max': [max_JP_NES, max_JP_WII],
    'Ventas NA Min': [min_NA_NES, min_NA_WII],
    'Ventas NA Max': [max_NA_NES, max_NA_WII]
}

df_ventas = pd.DataFrame(datos)
df_ventas
```

```python
# MEDIA DE VENTAS JA y NA por NES Y WII

media_NES_JA = nes["Ventas JP"].mean()
media_WII_JA = wii["Ventas JP"].mean()
media_NES_NA = nes["Ventas NA"].mean()
media_WII_NA = wii["Ventas NA"].mean()
```

```python
# TABLA DE MEDIAS 

datos = {
    'Plataforma': ['NES', 'WII'],
    'Media JP': [media_NES_JA, media_WII_JA],
    'Media NA': [media_NES_NA, media_WII_NA]
}

df_medias = pd.DataFrame(datos)
df_medias
```

```python
# TABLA DE MEDIA DE VENTAS NA Y JP

ventas_NA_JA = pd.DataFrame({
    'medias': [media_NES_JA, media_WII_JA, media_NES_NA, media_WII_NA]
}, index= ["media_NES_JA","media_WII_JA","media_NES_NA","media_WII_NA"])

ventas_NA_JA
```

```python
# SUMA DE MEDIAS

suma_medias = media_NES_JA + media_WII_JA + media_NES_NA + media_WII_NA
suma_medias
```

```python
# GRAFICO DE MEDIAS

etiquetas = ["media_NES_JA","media_WII_JA","media_NES_NA","media_WII_NA"]
ventas_NA_JA.plot(kind="pie", subplots=True, labels=etiquetas, autopct='%1.1f%%', figsize=(6,8))
plt.title('Distribución de Ventas por sus medias en JA Y NA');
```

```python
# MEDIANA JP Y NA POR WII Y NES

mediana_NES_NA = nes["Ventas NA"].median()
mediana_WII_NA = wii["Ventas NA"].median()
mediana_NES_JA = nes["Ventas JP"].median()
mediana_WII_JA = wii["Ventas JP"].median()

datos = {
    "Consola/region": ["NES NA", "WII NA", "NES JP", "WII JP"],
    "Mediana Ventas": [mediana_NES_NA, mediana_WII_NA, mediana_NES_JA, mediana_WII_JA]
}

resultados_df = pd.DataFrame(datos)
resultados_df
```

```python
# MODAS JP Y NA POR WII Y NES

moda_NES_NA = nes["Ventas NA"].mode()
moda_WII_NA = wii["Ventas NA"].mode()
moda_NES_JA = nes["Ventas JP"].mode()
moda_WII_JA = wii["Ventas JP"].mode()
```

```python
# LISTA DE MODAS

moda_NES_NA, moda_WII_NA, moda_NES_JA, moda_WII_JA
```

```python
# MEDIA DE VENTAS EU Y OTROS

media_eu = df["Ventas EU"].mean()
media_otros = df["Ventas Otros"].mean()

datos = {
    "Region": ["EU", "OTROS"],
    "Media Ventas": [media_eu, media_otros]
}

resultados_df = pd.DataFrame(datos)
resultados_df
```

```python
# TABLA DE VENTAS GLOBALES POR PLATAFORMA

df_plataforma = df.groupby("Plataforma")["Ventas Global"].sum().sort_values(ascending = False)
df_plataforma_por = df_plataforma / df["Ventas Global"].sum() * 100
df_plataforma_por
```

```python
# GRÁFICO VENTAS GLOBALES POR PLATAFORMA

df_plataforma_por.plot(kind = "pie", autopct='%1.1f%%', figsize=(11,11))
plt.title("% De ventas globales de juegos por plataforma");
```

```python
# GRÁFICO VENTAS GLOBALES POR PLATAFORMA

ax = df_plataforma_por.plot(kind="bar", figsize=(15, 4))
plt.title("% De ventas globales de juegos por plataforma")
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}%',  
                (p.get_x() + p.get_width() / 2., p.get_height()),  
                ha='center', va='center',  
                fontsize=7, color='black',  
                xytext=(0, 8), textcoords='offset points');
```
