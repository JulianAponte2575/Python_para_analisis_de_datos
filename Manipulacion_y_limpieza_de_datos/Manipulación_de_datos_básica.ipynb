# EJERCICIO BÁSICO DE MANIPULACIÓN DE DATOS EN PYTHON 

Ejercicio básico para manipular datos con pandas para extraer determinadas columnas y filas que cumplan con las condiciones.

```python
# IMPORTACIÓN DE LIBRERÍAS

import pandas as pd
```

```python
# DEFINICIÓN DE RUTA

ruta = "C:/Users/Julian/Documents/Ejecicios del botcamp/titanic.csv"
```

```python
# LECTURA DEL ARCHIVO Y VISUALIZACIÓN 

titanic = pd.read_csv(ruta)
titanic
```

```python
# PROMEDIO DE EDAD DE LAS PERSONAS EN EL BARCO

promedio_edad = titanic["Age"].mean()
promedio_edad
```

```python
# DATAFRAME DE SOLO LOS HOMBRES CON COLUMNAS ESPECIFICAS

df_hombres = pd.DataFrame(titanic[titanic["Sex"] == "male"],columns=["Name","Sex","Age","Ticket"])
df_hombres.reset_index(drop = True) # Esto reinicia el indice
```

```python
# DATAFRAME DE SOLO LOS MUJERES CON COLUMNAS ESPECIFICAS

df_mujeres = pd.DataFrame(titanic[titanic["Sex"] == "female"],columns=["Name","Age","Sex","Ticket"])
df_mujeres.reset_index(drop = True)
```

```python
# DATAFRAME CON NOMBRES Y EDAD DE LOS MAYORES DE 30 

mayores_a_30 = pd.DataFrame(titanic[titanic["Age"] > 30],columns=["Name","Age"])
mayores_a_30.reset_index(drop = True)
```
