## EJERCICIO DE MANIPULACIÓN DE DATOS

Este ejercicio de manipulación de datos básica, muestra una forma sencilla de como aprender a manejar la librería de pandas.  

```python
# IMPORTAR LIBRERÍA 

import pandas as pd
```

```python
# GENERAR UN DATAFRAME PERO SIN DATOS SIMULADOS 

data = {
    'Producto': ['Laptop', 'Tablet', 'Smartphone', 'Laptop', 'Tablet', 'Smartphone', 'Laptop', 'Tablet', 'Smartphone', 'Laptop'],
    'Precio': [1200, 450, 800, 1200, None, 850, 1150, 480, 800, None],
    'Cantidad': [5, 10, 15, None, 7, 8, 10, 5, None, 3],
    'Fecha de Venta': ['2023-07-01', '2023-07-02', '2023-07-03', '2023-07-04', '2023-07-05', '2023-07-06', '2023-07-07', '2023-07-08', '2023-07-09', '2023-07-10'],
    'Ciudad': ['Bogotá', 'Medellín', 'Bogotá', 'Cali', 'Bogotá', 'Medellín', 'Cali', None, 'Bogotá', 'Cali']
}

df = pd.DataFrame(data)
df
```

```python
# VER INFORMACIÓN DE COLUMNAS 

df.info()
```

```python
# CONTAR DATOS NULOS 

df.isna().sum().sort_values(ascending = False)
```

```python
# IMPUTAR DATOS NULOS DE LA COLUMNA PRECIO CON LA MEDIANA

df["Precio"] = df["Precio"].fillna(df["Precio"].median())
df
```

```python
# IMPUTAR DATOS NULOS DE LA COLUMNA CANTIDAD CON LA MEDIANA

df["Cantidad"] = df["Cantidad"].fillna(df["Cantidad"].median())
df
```

```python
# IMPUTAR DATOS NULOS DE LA COLUMNA CIUDAD CON UNA CATEGORIA DSTINTA

df["Ciudad"] = df["Ciudad"].fillna("Desconocido")
df
```

```python
# CAMBIAR FORMATOS DE COLUMNAS DE FLOTANTES A ENTEROS

df["Precio"] = df["Precio"].astype(int)
df["Cantidad"] = df["Cantidad"].astype(int)
df
```

```python
# CAMBIAR FORMATO TEXTO DE FECHA A TIPO FECHA

df["Fecha de Venta"] = pd.to_datetime(df["Fecha de Venta"], format = "%Y-%m-%d")
df
```

### CÁLCULOS BÁSICOS

Cálculos básicos con las columnas del dataframe.

```python
# CALCULAR INGRESOS POR CIUDAD

df["ingresos"] = df["Cantidad"] * df["Precio"]
ingresos_por_ciudad = df.groupby("Ciudad")["ingresos"].sum()
valores = {"Ciudad": ingresos_por_ciudad.index, "Ingresos": ingresos_por_ciudad.values}
ingresos_por_ciudad = pd.DataFrame(valores)
ingresos_por_ciudad
```

```python
# MOSTRAR EL PRODUCTO CON MAYORES INGRESOS

producto_mayor_ingreso = df.groupby("Producto")["ingresos"].sum().sort_values(ascending = False).head(1)
producto_mayor_ingreso
```

```python
# CALCULAR LA CIUDAD CON MAYOR UNIDADES VENDIDAS

df["ingresos"] = df["Cantidad"] * df["Precio"]
ciudad_con_mayores_ventas = df.groupby("Ciudad")["Cantidad"].sum().sort_values(ascending = False).head(1)
ciudad_con_mayores_ventas
```
