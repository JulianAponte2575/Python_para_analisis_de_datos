## PROYECTO DE ANÁLISIS DE DATOS CON BASE DE DATOS DE DÉLITOS DE LOS ÁNGELES

```python
# IMPORTAR LIBRERIAS

import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
# TRAER EL DATASET

ruta = "C:/Users/Julian/Downloads/Crime_Data_from_2020_to_Present.csv/Crimenes.csv"
df = pd.read_csv(ruta)
df
```

```python
# VER VALORES ÚNICOS POR ÁREA

unicos_area = df["AREA"].unique()
unicos_area
```

```python
# INFORMACIÓN DE LA TABLA

df.info()
```

```python
# CRÍMENES POR SEXO DE LA VICTIMA

sexo = df.groupby("SEXO_VICTIMA")["SEXO_VICTIMA"].count().sort_values(ascending = False)
sexo
```

```python
# RELLENAR NULOS CON X

df["SEXO_VICTIMA"] = df["SEXO_VICTIMA"].fillna("X")
```

```python
# RELLENAR NULOS CON "Desconocido"

df["LUGAR_CRIMEN"] = df["LUGAR_CRIMEN"].fillna("Desconocido")
```

```python
# REEMPLAZAR SEXO CON X

df["SEXO_VICTIMA"] = df["SEXO_VICTIMA"].replace("H","X")
df["SEXO_VICTIMA"] = df["SEXO_VICTIMA"].replace("-","X")
```

```python
# VALIDAR INFORMACIÓN DE LA TABLA

df.info()
```

```python
# ESTANDARIZAR FECHA

df["FECHA_DELITO"] = pd.to_datetime(df["FECHA_DELITO"], format='mixed')
df
```

```python
# CONVERTIR HORA DE DELITO EXTRAYENDO LA HORA 

df['HORA_DELITO'] = pd.to_datetime(df['HORA_DELITO'].astype(str).str.zfill(4), format='%H%M').dt.hour
df
```

```python
# DELITOS POR ÁREA

areas = df.groupby("AREA")["AREA"].count().sort_values(ascending = False)
por_areas = (areas / len(df) * 100).round(2).head(6)
por_areas
```

```python
# PORCENTAJE DE DELITOS

delitos = df.groupby("DELITO")["DELITO"].count().sort_values(ascending = False)
por_delitos =  (delitos / len(df) * 100).round(2).head(6)
por_delitos
```

```python
areas_delitos = df.groupby(["AREA", "DELITO"])["AREA"].count().sort_values(ascending = False)
areas_delitos
```

```python
# CRIMENES POR LUGAR

lugar = df.groupby("LUGAR_CRIMEN")["LUGAR_CRIMEN"].count().sort_values(ascending = False)
por_lugar = (lugar / len(df) * 100).round(2).head(6)
por_lugar
```

```python
# AGREGAR COLUMNAS DE MES Y DIA

df_por_mes = df.copy()
df_por_mes["FECHA_DELITO"] = pd.to_datetime(df_por_mes["FECHA_DELITO"])
df_por_mes["MES_DELITO"] = df_por_mes["FECHA_DELITO"].dt.month_name()
df_por_mes['DIA_DELITO_NOMBRE'] = df_por_mes['FECHA_DELITO'].dt.day_name()
df_por_mes
```

```python
mes_delito = df_por_mes.groupby("MES_DELITO")["MES_DELITO"].count().sort_values(ascending = False)
por_mes = mes_delito / len(df_por_mes) * 100
por_mes
```

```python
dia_delito = df_por_mes.groupby("DIA_DELITO_NOMBRE")["DIA_DELITO_NOMBRE"].count().sort_values(ascending = False)
por_dia = (dia_delito / len(df_por_mes) * 100).round(2)
por_dia
```

```python
hora_delito = df_por_mes.groupby("HORA_DELITO")["HORA_DELITO"].count().sort_values(ascending = False)
porcentaje_hora = (hora_delito / len(df_por_mes) * 100).round(2)
porcentaje_hora 
```

```python
# TIPO DE DELITO POR DÍA

delitos_por_dia = df_por_mes.groupby(["DELITO", "DIA_DELITO_NOMBRE"]).size().unstack(fill_value=0)
delitos_por_dia.sort_values(by="Friday", ascending= False)
```

```python
df_ml = df_por_mes.copy()
df_ml = df_ml.drop(columns=["FECHA_DELITO", "ESTADO","LAT","LON"], axis = 1)
df_ml
```

```python
# EXTRAER COLUMNAS NÚMERICAS

num = df_ml.select_dtypes('number').reset_index(drop = True)
```

```python
# CONVERTIR VARIABLES CATEGORICAS A NÚMERICAS 

from sklearn.preprocessing import OneHotEncoder

#Categóricas
cat = df_ml.select_dtypes('O')

#Instanciamos
ohe = OneHotEncoder(sparse_output = False)

#Entrenamos
ohe.fit(cat)

#Aplicamos
cat_ohe = ohe.transform(cat)

#Ponemos los nombres
cat_ohe = pd.DataFrame(cat_ohe, columns = ohe.get_feature_names_out(input_features = cat.columns)).reset_index(drop = True)
```

```python
# UNIR NUMERICAS Y CATEGORICAS

df_transformado = pd.concat([cat_ohe,num], axis = 1)
df_transformado
```

```python
# LIBRERIAS PARA EL MAPA DE CALOR

import folium
from folium.plugins import HeatMap
```

```python
# Filtramos las filas que tienen valores nulos en LAT y LON, ya que no pueden ser usadas en el mapa de calor
df = df.dropna(subset=['LAT', 'LON'])

# Inicializa el mapa centrado en la ubicación general de Los Ángeles
heat_map = folium.Map(location=[34.05, -118.25], zoom_start=12)

# Prepara los datos para el HeatMap (latitud, longitud)
heat_data = [[row['LAT'], row['LON']] for index, row in df.iterrows()]

# Crea el mapa de calor y añádelo al mapa
HeatMap(heat_data).add_to(heat_map)

# Guarda el mapa en un archivo HTML
heat_map.save("heatmap_delitos.html")

print("Mapa de calor generado exitosamente y guardado como 'heatmap_delitos.html'")
```
