## VALOR DE VIDA DEL CLIENTE CON (ANÁLISIS DE DATOS)

```python
import pandas as pd
import random
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# CREACIÓN DE LOS DATOS 
n = 100
data = {
    "Cliente_ID": [i for i in range(1, n + 1)],
    "Nombre": [f"Cliente_{i}" for i in range(1, n + 1)],
    "Fecha_Registro": [datetime.now() - timedelta(days=random.randint(30, 1000)) for _ in range(n)],
    "Última_Compra": [datetime.now() - timedelta(days=random.randint(1, 30)) for _ in range(n)],
    "Frecuencia_Compra": [random.randint(1, 12) for _ in range(n)],
    "Monto_Total": [random.randint(200, 3000) for _ in range(n)],
    "Costo_Adquisición": [random.randint(50, 200) for _ in range(n)],
}
```

```python
# TABLA
df = pd.DataFrame(data)
df
```

### ANÁLISIS DE CLV DE CLIENTES 

```python
# CALCULO DE CLV 

df["CLV"] = (df["Frecuencia_Compra"] * df["Monto_Total"]) - df["Costo_Adquisición"]
df
```

```python
# AÑADIR AÑO DE ULTIMA COMPRA Y AÑO DE REGISTRO

df["Año"] = df["Última_Compra"].dt.year
df["Año_Registro"] = df["Fecha_Registro"].dt.year
df
```

```python
# TASA DE RETENCIÓN DE CLIENTES EN EL 2024  

datos_2024 = df[df["Año"] == 2024]
clientes_repetidos = datos_2024[datos_2024['Frecuencia_Compra'] > 1]
num_clientes_repetidos = clientes_repetidos['Cliente_ID'].count()
num_clientes_repetidos / len(df)
```

```python
# PROMEDIO DE COSTO DE VIDA POR AÑO

clv_año_registro = df.groupby("Año_Registro")["CLV"].mean()
clv_año_registro
```

```python
# DISTRIBUCIÓN CLV POR AÑO

sns.set_style("whitegrid", {'axes.grid': False})
g = sns.displot(data=df,
           x="CLV",
           col="Año",
           kde= True, 
            bins = 8)
g.fig.suptitle("Distribución del CLV", 
               fontsize=16, 
               ha='center')
g.fig.subplots_adjust(top=0.85)
```

```python
# DISTRIBUCIÓN COSTO DE ADQUISIÓN 

sns.set_style("whitegrid", {'axes.grid': False})
g = sns.displot(data=df,
           x="Costo_Adquisición",
           col="Año",
           kde= True, 
            bins = 8)
g.fig.suptitle("Distribución del Costo de adquisión", 
               fontsize=16, 
               ha='center')
g.fig.subplots_adjust(top=0.85)
```

```python
# CORRELACIÓN ENTRE EL COSTO DE ADQUISICIÓN Y EL CLV

sns.set_style("whitegrid", {'axes.grid': False})
g = sns.relplot(data=df,
           x="Costo_Adquisición",
           y="CLV",
           col="Año")
g.fig.suptitle("Relación entre Costo de adquisición y el CLV", 
               fontsize=16, 
               ha='center')
g.fig.subplots_adjust(top=0.85);
```

## SEGMENTACIÓN DE CLIENTES SEGÚN SU CLV

```python
# CREACIÓN DE LA COLUMNA SEGMENTO DE ACUERDO AL CLV 

limites = [0, 8000, 18000, float("inf")]
segmentos = ["Bajo","Medio","Alto"]

df["Segmento"] = pd.cut(df['CLV'], bins=limites, labels=segmentos, right=False)
df
```

## ANÁLISIS POR SEGMENTO DE TODOS LOS CLIENTES

```python
# FRECUENCIA DE COMPRA POR SEGMENTO Y AÑO

frec = pd.crosstab(df["Año"],df["Segmento"],aggfunc='mean',values=df["Frecuencia_Compra"])
frec
```

```python
# COSTO DE ADQUISICIÓN POR SEGMENTO y AÑO 

ADQ = pd.crosstab(df["Año"],df["Segmento"],aggfunc='mean',values=df["Costo_Adquisición"])
ADQ
```

```python
# VALOR DE VIDA PROMEDIO POR SEGMENTO 

CLV_segmento = df.groupby("Segmento", observed=True)["CLV"].mean()
CLV_segmento
```

```python
# COSTO DE ADQUISIÓN POR SEGMENTO
sns.set_style("whitegrid", {'axes.grid': False})
sns.catplot(data=df,
           kind="box",
           x="Segmento",
            y="Costo_Adquisición")
plt.title("Costo de adquisición por segmento");
```

```python
# DISTRIBUCIÓN CLV POR SEGMENTO 

sns.set_style("whitegrid", {'axes.grid': False})
g = sns.displot(data=df,
           x="CLV",
           col="Segmento",
           kde= True, 
            bins = 8)
g.fig.suptitle("CLV distribución por segmento", 
               fontsize=16, 
               ha='center')
g.fig.subplots_adjust(top=0.85);
```

```python
# CORRELACIÓN ENTRE FRECUENCIA DE COMPRA Y COSTO DE ADQUISICIÓN 
sns.set_style("whitegrid", {'axes.grid': False})
g = sns.relplot(data=df,
           x="Frecuencia_Compra",
           y="Costo_Adquisición",
           col="Segmento")
g.fig.suptitle("Relación entre Frecuencia de Compra y costo de adquisión por Segmento", 
               fontsize=16, 
               ha='center')
g.fig.subplots_adjust(top=0.85);
```

```python
# GRAFICO DE BIGOTES CLV POR AÑO POR SEGMENTO 
sns.set_style("whitegrid", {'axes.grid': False})
sns.catplot(data=df,
           kind="box",
           x="Año_Registro",
           y="CLV",
           col = "Segmento")
plt.title("CLV por año");
```

```python
# CORRELACIÓN ENTRE EL COSTO DE ADQUISICIÓN Y EL CLV POR AÑO 
sns.set_style("whitegrid", {'axes.grid': False})
sns.relplot(data=df,
           x="Costo_Adquisición",
           y="CLV",
           col = "Segmento")
plt.title("CLV vs costo de adquisión");
```

```python
# CORRELACIÓN ENTRE FRECUENCIA DE COMPRA Y EL CLV POR AÑO 
sns.set_style("whitegrid", {'axes.grid': False})
sns.relplot(data=df,
           x="Frecuencia_Compra",
           y="CLV",
           col = "Segmento")
plt.title("CLV vs costo de adquisión");
```

```python
# TABLA DE CORRELACIÓN 

corre = df[["Frecuencia_Compra","Monto_Total","Costo_Adquisición","CLV"]].corr()
corre
```
